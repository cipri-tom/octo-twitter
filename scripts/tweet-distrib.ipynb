```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import pandas as pd
```

```python
files = ['../data/train_neg_full_preprocess.txt', '../data/train_pos_full_preprocess.txt']
num_tweets = 0
counts = defaultdict(int)
for path in files:
    with open(path, 'r') as file:
        for tweet in file:
            l = len(tweet.split(' '))
            counts[l]  += 1
            num_tweets += 1
```

```python
x = np.arange(max(counts.keys()))
y = [counts[i] for i in x]
```

```python
plt.figure(figsize=(12,6))
plt.bar(x, y, align='center')
plt.xlabel("Number of words")
plt.ylabel("Number of tweets")
plt.title("Train set (Positive & Negative)")
```

```python
plt.savefig('train_distrib.png', bbox_inches="tight")
```
